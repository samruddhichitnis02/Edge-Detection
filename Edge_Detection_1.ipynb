**Edge Detection using Pillow**

Edge Detection internally works by running a filter/Kernel over a Digital Image

There are two forms of edge detection:

Search Based Edge detection (First order derivative)

Zero Crossing Based Edge detection (Second order derivative)


Some of the commonly known edge detection methods are:

Laplacian Operator or Laplacian Based Edge detection (Second order derivative)

Canny edge detector (Second order derivative)

Prewitt operator (First order derivative)

Sobel Operator (First order derivative)

```
from PIL import Image, ImageFilter 
import matplotlib.pyplot as plt
```

```
img = Image.open('/content/drive/My Drive/Colab Notebooks/Dog.jpg')
```

```
plt.imshow(img)
plt.show()
```

Convert the image to gray scale as the edge detection requires input image to be of mode = Greyscale (L) 

```
img = img.convert("L") 
```

If you have an L mode image, that means it is a single channel image - normally interpreted as greyscale. The L means that is just stores the Luminance. It is very compact, but only stores a greyscale, not colour.

```
plt.imshow(img)
plt.show()
```

**Method1**

Detecting Edges on the Image using the argument ImageFilter.FIND_EDGES

```
image = img.filter(ImageFilter.FIND_EDGES) 
```

```
plt.imshow(image)
plt.show()
```

**Method2**

```
final = img.filter(ImageFilter.Kernel((3, 3), (-1, -1, -1, -1, 8, 
                                          -1, -1, -1, -1), 1, 0)) 
```

```
plt.imshow(final)
plt.show()
```

Here the advantage of using a kernel instead of using inbuit function is that we can define kernels according to our use
